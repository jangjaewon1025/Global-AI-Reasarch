```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train = pd.read_csv('../data/sleep_X_train.csv')
X_test = pd.read_csv('../data/sleep_X_test.csv')
y_train = pd.read_csv('../data/sleep_y_train.csv')
y_test = pd.read_csv('../data/sleep_y_test.csv')
```

```python
y_train.head()
```

```python
X_train = X_train.loc[:,'WorkoutTime':]
X_test = X_test.loc[:,'WorkoutTime':]
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train['SleepTime'])
```

```python
model.coef_
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(y_train['SleepTime'], y_train_pred, label='training', color='lightblue', s=100, alpha=0.8, marker='o')
plt.scatter(y_test['SleepTime'], y_test_pred, label='test', color='blue', s=100, alpha=0.8, marker='o')

# Add plot labels and legend
plt.title('SleepTime Prediction', fontsize=16)
plt.xlabel('SleepTime', fontsize=14)
plt.ylabel('Model', fontsize=14)
plt.legend(title='Data Type', fontsize=12, title_fontsize=13)
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
from sklearn.linear_model import Ridge
# Create models
ridge = Ridge(
    # Regularization penalty
    alpha=10,
    random_state=1)
# Fit object
ridge.fit(X_train, y_train)
```

```python
y_train_pred_ridge = ridge.predict(X_train)
y_test_pred_ridge = ridge.predict(X_test)
```

```python
# Evaluate model
print(f'Training R^2: {ridge.score(X_train, y_train)}')
print(f'Test R^2: {ridge.score(X_test, y_test)}')
print(f'Train RMSE: {mean_squared_error(y_train, y_train_pred_ridge, squared=False)}')
print(f'Test RMSE: {mean_squared_error(y_test, y_test_pred_ridge, squared=False)}')
```
